```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('mysolutionsdatas.csv')
df
```

```python
fig = plt.figure(figsize=(15,9))
ax = fig.add_subplot()
Unemployment_Rate = [9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]
  
ax.plot(df.dfs_solutions*1000,df.index, color='red',label = 'Depth-first search')
ax.plot(df.bfs_solutions*1000,df.index, color='blue',label = 'Breadth-first search')

ax.set_xlabel('Time spent (ms)', fontsize=20,font = 'Times new roman')
ax.set_ylabel('Solutions found',fontsize=20,font = 'Times new roman')

ax.set_yticks(range(0,93,10))
ax.set_xticks(range(0,int(df.bfs_solutions.max()*1000),10),fontsize= 18)
plt.yticks(fontsize = 16,font = 'Times New Roman')
plt.xticks(fontsize = 16,font = 'Times New Roman')

ax.legend(fontsize= 16)
ax.grid()
```
